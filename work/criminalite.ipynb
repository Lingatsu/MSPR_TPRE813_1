# Data delinquance

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Delinquance

```python
path="./data/criminalite/donnee-comm-data.gouv-parquet-2023-geographie2024-produit-le2024-07-05.parquet"
delinquance_brut = pd.read_parquet(path, engine='pyarrow')
delinquance_brut.tail()
```

```python
delinquance_drop = delinquance_brut.drop(columns=['millPOP','millLOG'])
delinquance_drop
```

```python
delinquance_rename = delinquance_drop.rename(columns={'CODGEO_2024' : 'CODGEO'})
delinquance_rename
```

## Informations complémentaires

```python
path2="./data/criminalite/info-complements-data.gouv-2023-geographie2024-produit-le2024-07-05.xlsx"
complement = pd.read_excel(path2)
complement.head()
```

```python
complement_crop = complement[['CODGEO','LIBGEO']]
complement_crop
```

## Concaténation

```python
concatenation=pd.concat([delinquance_rename, complement_crop])
concatenation_rename = concatenation.rename(columns={'CODGEO' : 'codecommune','annee' : 'year'})
concatenation_rename
```

```python
concatenation_drop = concatenation_rename.drop(columns=['LIBGEO'])
concatenation_duplicate = concatenation_drop.dropna(subset=['year'])
concatenation_duplicate
```

```python
# conversion de string à int avec regex pour eviter les problèmes de changement de valeurs de base 10 (ex: 2A001 au lieu de 1001 par exemple)
concatenation_duplicate['codecommune'] = concatenation_duplicate['codecommune'].str.replace('[^0-9]', '', regex=True).astype('int64')

# retire les premiers 0 pour mettre au même format que les DF de mes camarades
concatenation_duplicate['year'] = concatenation_duplicate['year'].astype(str)
concatenation_duplicate['year'] = concatenation_duplicate['year'].str.replace('.0','')
concatenation_duplicate['year'] = '20' + concatenation_duplicate['year']
concatenation_duplicate['year'] = concatenation_duplicate['year'].str.replace('[^0-9]', '', regex=True).astype('int64')
concatenation_duplicate
```
