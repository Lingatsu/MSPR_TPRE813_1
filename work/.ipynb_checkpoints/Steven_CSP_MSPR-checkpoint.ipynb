```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re

from google.colab import drive
drive.mount('/content/drive')
```

```python
pathCSPCommune = '/content/drive/MyDrive/EPSI/MSPR/CSP/cspcommunes.csv'

dfCSPCommune = pd.read_csv(pathCSPCommune)

# pd.set_option('display.max_columns', None)
```

```python
new_column_names = [re.sub(r'\d', '', col) for col in dfCSPCommune.columns]

unique_column_names = set(new_column_names)

print(unique_column_names)
```

```python
def filter_columns(df):
  keywords = ['dep', 'nomdep', 'codecommune', 'nomcommune', 'agri', 'indp', 'cadr', 'pint', 'empl', 'ouvr', 'chom', 'pact']
  pattern = r"^(dep|nomdep|codecommune|nomcommune|agri|indp|cadr|pint|empl|ouvr|chom|pact)(\d{4})?$"
  filtered_columns = [col for col in df.columns if re.match(pattern, col)]
  df_filtered = df[filtered_columns]
  return df_filtered

def reorder_columns(df):
    first_columns = ['dep', 'nomdep', 'codecommune', 'nomcommune']

    remaining_columns = [col for col in df.columns if col not in first_columns]

    remaining_columns_sorted = sorted(remaining_columns)

    ordered_columns = first_columns + remaining_columns_sorted

    df_reordered = df[ordered_columns]

    return df_reordered

df_filtered = filter_columns(dfCSPCommune)

df_reordered = reorder_columns(df_filtered)

print(df_reordered.head())
print(df_reordered.columns.tolist())
```

```python
def pivot_to_long_format(df):
  id_vars = ['dep', 'nomdep', 'codecommune', 'nomcommune']
  value_vars = [col for col in df.columns if re.match(r'(agri|indp|cadr|pint|empl|ouvr|chom|pact)\d{4}', col)]

  df_melted = pd.melt(df, id_vars, value_vars=value_vars, var_name = 'variable', value_name = 'value')

  df_melted['année'] = df_melted['variable'].str.extract(r'(\d{4})').astype(int)
  df_melted['category'] = df_melted['variable'].str.extract(r'(agri|indp|cadr|pint|empl|ouvr|chom|pact)')

  df_pivot = df_melted.pivot_table(index=['dep', 'nomdep', 'codecommune', 'nomcommune', 'année'], columns='category', values='value').reset_index()

  final_columns = ['dep', 'nomdep', 'codecommune', 'nomcommune', 'année', 'agri', 'indp', 'cadr', 'pint', 'empl', 'ouvr', 'chom', 'pact']

  df_pivot = df_pivot.reindex(columns=final_columns)

  return df_pivot

df_long_format = pivot_to_long_format(df_reordered)

print(df_long_format.head())
```
